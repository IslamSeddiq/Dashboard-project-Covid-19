**Preprocessing(Data cleaning)**

```python
df.drop(['City or Regency', 'Special Status', 'Country', 'Continent', 'Time Zone'], axis=1, inplace=True)
```

```python
df.columns
```

**Fixing data types**

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# removing the % sign so that we can convert the data type to float
df['Case Fatality Rate'] = df['Case Fatality Rate'].str.rstrip('%').astype(float)
df['Case Recovered Rate'] = df['Case Recovered Rate'].str.rstrip('%').astype(float)
```

```python
pd.DataFrame({'DataType': df[['Date', 'Case Fatality Rate', 'Case Recovered Rate']].dtypes})
```

**Handling the missing values**

```python
# Province column
df['Province'].fillna('Country Level', inplace=True)
# Island column
df['Island'].fillna('Country Level', inplace=True)
# Total Cities
df['Total Cities'].fillna(0, inplace=True)
# Total Urban and Rural Villages &
df['Total Urban Villages'].fillna(0, inplace=True)
df['Total Rural Villages'].fillna(0, inplace=True)
```

-----> Handling the missing values of the 'Growth Factor of New Cases' and 'Growth Factor of New Deaths' columns by calculating them.

```python
import numpy as np

# Boolean mask for missing values
missing = df['Growth Factor of New Cases'].isna()

# Apply logic only to missing rows
df.loc[missing, 'Growth Factor of New Cases'] = np.where(
    df['New Cases'][missing] == df['New Cases'].shift(1)[missing],
    1,
    df['New Cases'][missing] / df['New Cases'].shift(1)[missing]
)
```

```python
missing = df['Growth Factor of New Deaths'].isna()

df.loc[missing, 'Growth Factor of New Deaths'] = np.where(
    df['New Deaths'][missing] == df['New Deaths'].shift(1)[missing],
    1,
    df['New Deaths'][missing] / df['New Deaths'].shift(1)[missing]
)
```

-----> This way does not handle all missing values because of the infinite
values that we are going to convert them to NaN, but at least it is going to reduce them.

```python
# converting the inf values to NaN
# This is helpful to clean our data before visualization
df['Growth Factor of New Cases'].replace([np.inf, -np.inf], np.nan, inplace=True)
df['Growth Factor of New Deaths'].replace([np.inf, -np.inf], np.nan, inplace=True)
```

**NOTE**: We have noticed that the total recovered patients greater than the total cases so that the 'Total Active Cases' column contains negative values which is not logic, so we need to restate the real values of these columns.

```python
# check calculation
# Find the condition where Total Active Cases is negative
negative_values = df['Total Active Cases'] < 0

# Swap 'Total Cases' and 'Total Recovered' only for those rows
df.loc[negative_values, ['Total Cases', 'Total Recovered']] = df.loc[negative_values, ['Total Recovered', 'Total Cases']].values
```

```python
df[['Total Cases', 'Total Recovered', 'Total Active Cases']]
```

```python
# recalculate the 'Total Active Cases' column
df['Total Active Cases'] = df['Total Cases'] - (df['Total Recovered'] + df['Total Deaths'])
df['Total Active Cases']
```

**Visualization**

```python
df.groupby('Date')['New Cases'].sum().plot(kind='line', color='blue', figsize=(12, 5))
plt.title('Daily COVID-19 New Cases')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.groupby('Province')['Total Cases'].max().sort_values().plot(kind='bar', color='orange', figsize=(10, 5))
plt.title('Total Cases by Province')
plt.grid(True)
plt.show()
```

```python
df[['Total Recovered', 'Total Deaths', 'Total Active Cases']].median().plot(kind='bar', color=['green', 'red', 'blue'], figsize=(7, 5))
plt.title('Compare the status of outcomes with median')
plt.show()
```

```python
df[['New Cases', 'Growth Factor of New Cases']].plot.scatter(x='New Cases', y='Growth Factor of New Cases', color='blue', figsize=(7, 5))
plt.title('Relationship between New Cases and Growth Factor of New Cases')
plt.xlabel('New Cases')
plt.ylabel('Growth Factor')
plt.show()
```
